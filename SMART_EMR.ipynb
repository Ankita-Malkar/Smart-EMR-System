```python
from IPython.display import Javascript
from google.colab.output import eval_js
from base64 import b64decode
import io
import soundfile as sf

def record_audio(seconds=5):
    display(Javascript("""
    async function recordAudio(sec) {
        const stream = await navigator.mediaDevices.getUserMedia({audio: true});
        const recorder = new MediaRecorder(stream);
        const chunks = [];
        recorder.ondataavailable = e => chunks.push(e.data);
        recorder.start();
        await new Promise(resolve => setTimeout(resolve, sec * 1000));
        recorder.stop();
        return new Promise(resolve => {
            recorder.onstop = () => {
                const blob = new Blob(chunks);
                const reader = new FileReader();
                reader.onload = () => resolve(reader.result.split(',')[1]);
                reader.readAsDataURL(blob);
            };
        });
    }
    """))

    audio_data = eval_js(f"recordAudio({seconds})")
    audio_bytes = b64decode(audio_data)
    return audio_bytes
```

```python
from faster_whisper import WhisperModel

model = WhisperModel("medium", device="cuda", compute_type="float16")
```

```python
def transcribe_audio(audio_bytes):
    audio_file = io.BytesIO(audio_bytes)
    audio_array, sample_rate = sf.read(audio_file)

    segments, _ = model.transcribe(audio_array)

    transcript = ""
    for segment in segments:
        transcript += segment.text

    return transcript
```

```python
nlp = spacy.load("en_ner_bc5cdr_md")
```

```python
negex = medspacy.load()
nlp.add_pipe("negex")
```

```python
tokenizer = BertTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
clinical_model = BertModel.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
clinical_model.to(device)
clinical_model.eval()
```

```python
icd_df = pd.read_csv("icd10_codes.csv")
icd_df = icd_df.dropna()
icd_df.columns = ["code", "description"]
icd_df["description"] = icd_df["description"].str.lower()
```

```python
def compute_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=128)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = clinical_model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu()

icd_embeddings = []
for desc in icd_df["description"][:3000]:
    icd_embeddings.append(compute_embedding(desc))
```

```python
def normalize_text(text):
    text = text.lower()
    text = text.replace("over", "/")
    text = re.sub(r"\s+", " ", text)
    return text
```

```python
def extract_entities(text):

    text = normalize_text(text)
    doc = nlp(text)

    symptoms = []
    diseases = []
    medications = []
    allergies = []

    for ent in doc.ents:
        if ent._.negex:
            continue

        if ent.label_ == "DISEASE":
            diseases.append(ent.text)

        if ent.label_ == "CHEMICAL":
            medications.append(ent.text)

            if "allergic to" in text:
                allergies.append(ent.text)

    return list(set(diseases)), list(set(medications)), list(set(allergies)), doc
```

```python
def extract_duration(text):
    pattern = r'(\d+)\s*(days?|weeks?|months?|years?)'
    return re.findall(pattern, text)
```

```python
def extract_vitals(text):

    bp_pattern = r'(\d{2,3}/\d{2,3})'
    temp_pattern = r'(\d{2,3}\.?[\d]?\s?°?f)'

    return {
        "blood_pressure": re.findall(bp_pattern, text),
        "temperature": re.findall(temp_pattern, text)
    }
```

```python
def map_icd(text):

    text_embedding = compute_embedding(text)
    similarities = []

    for idx, icd_embed in enumerate(icd_embeddings):
        sim = cosine_similarity(
            text_embedding.numpy(),
            icd_embed.numpy()
        )[0][0]

        similarities.append((sim, icd_df.iloc[idx]["code"], icd_df.iloc[idx]["description"]))

    similarities.sort(reverse=True)

    return similarities[:5]
```

```python
def complete_emr_pipeline(transcript):

    transcript = normalize_text(transcript)

    diseases, medications, allergies, doc = extract_entities(transcript)

    durations = extract_duration(transcript)
    vitals = extract_vitals(transcript)
    icd_results = map_icd(transcript)

    emr_output = {
        "Symptom_or_Disease": diseases,
        "Duration": durations,
        "Vitals": vitals,
        "Medications": medications,
        "Allergies": allergies,
        "ICD_Mapping": icd_results
    }

    return emr_output
```

```python
def generate_doctor_document(transcript, emr_output):

    symptoms = emr_output["Symptom_or_Disease"]
    duration = emr_output["Duration"]
    vitals = emr_output["Vitals"]
    medications = emr_output["Medications"]
    allergies = emr_output["Allergies"]
    icd_results = emr_output["ICD_Mapping"]

    icd_codes = [f"{code} - {desc}" for _, code, desc in icd_results]

    doctor_doc = f"""
    ========================
    CLINICAL ENCOUNTER NOTE
    ========================

    Chief Complaint:
    {', '.join(symptoms) if symptoms else "Not specified"}

    Duration:
    {duration if duration else "Not specified"}

    Vital Signs:
    Blood Pressure: {vitals.get('blood_pressure', [])}

    Current Medications:
    {', '.join(medications) if medications else "None reported"}

    Allergies:
    {', '.join(allergies) if allergies else "None reported"}

    ICD Mapping:
    {chr(10).join(icd_codes)}
    """

    return doctor_doc
```

```python
def generate_patient_document(emr_output):

    symptoms = emr_output["Symptom_or_Disease"]
    duration = emr_output["Duration"]
    medications = emr_output["Medications"]

    diagnosis = emr_output["ICD_Mapping"][0][2] if emr_output["ICD_Mapping"] else "a health condition"

    patient_doc = f"""
    ========================
    PATIENT VISIT SUMMARY
    ========================

    You visited the doctor because of:
    {', '.join(symptoms) if symptoms else "your symptoms"}

    You have been experiencing this for:
    {duration if duration else "some time"}

    Based on the consultation, you may have:
    {diagnosis}

    Medications advised:
    {', '.join(medications) if medications else "As prescribed by your doctor"}

    Please follow the medical advice carefully.
    """

    return patient_doc
```
